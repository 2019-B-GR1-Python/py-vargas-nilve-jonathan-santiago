### Analizamos como los precios de los Iphones van cambiando con el transcurso del tiempo

### Analizamos los pesos en gramos de los Iphones. El iphone 11 pro es MAX  es el dispositivo que mas pesa con un total de 226 gramos.

```python
fig, ax = plt.subplots(figsize=(6, 8), subplot_kw=dict(aspect="equal"))

recipe = modelo_i
data = df_Iphone['PESO (g)']


wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5),labels=data, startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.5*np.sign(x), 1.9*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Peso (g) de los Iphones\n\n\n\n\n\n")

plt.show()
```

### Analizamos los pesos en gramos de los Samsung. El dispositivo movil mas pesado de esta marca es el Galaxy S9+ con 189 gramos. 

```python
fig, ax = plt.subplots(figsize=(6, 8), subplot_kw=dict(aspect="equal"))

recipe = modelo_s
data = df_Samsung['PESO (g)']


wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5),labels=data, startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.5*np.sign(x), 1.6*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Peso (g) de los Samsung\n\n\n\n\n\n")

plt.show()
```

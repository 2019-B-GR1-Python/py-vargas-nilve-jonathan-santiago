```python
#Imagenes como arreglos
#las imagenes que pueden pixelarse y las que no
import numpy as np
from scipy import ndimage
from scipy import misc
```

```python
mapache = misc.face()
#dimension
mapache.shape
```

```python
#que tipo de dato es mapache
print(type(mapache))
```

```python
#dibujar al mapache
import matplotlib.pyplot as plt

plt.imshow(mapache)

```

```python
mapache[::-1]
```

```python
# dar la vuelta a la imagen. porque la imagen es un arreglo de 3 dimensiones.
plt.imshow(mapache[::-1])
```

```python
#dividir en dos la imagen
arr1 , arr2, arr3, arr4 = np.hsplit(mapache,4)
#print(plt.imshow(arr1))

arrsub1, arrsub2 = np.hsplit(arr1,2)
print(plt.imshow(arrsub1))
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
class clase_rompecabezas():
    global rompecabezas_array
    global comprobar_pieza
    global comprobar_posicion

   
    def mostrar_rompecabezas():
        global groot 
        groot = np.asarray(Image.open("./minio.jpg"))  #transformar imagen a matriz numpy
        plt.imshow(groot);
        plt.title("MINIONS")
        plt.axis('off')
        plt.show()
        
        
   
    def piezas():
        global piezas_totales
            
        arr1 , arr2, arr3 = np.hsplit(groot,3)
    
        pieza_1, pieza_2 = np.vsplit(arr1,2)
        pieza_3, pieza_4 = np.vsplit(arr2,2)
        pieza_5, pieza_6 = np.vsplit(arr3,2)
        
        
        piezas_totales = [pieza_1, pieza_2,pieza_3, pieza_4,pieza_5, pieza_6]
        np.random.shuffle(piezas_totales)
        
        
        aux = 0
        
        for pieza in piezas_totales:
            plt.imshow(pieza)
            aux = aux + 1
            plt.title(f"Pieza {aux}")
            plt.show()
            
        
    def mostrar_tablero():
        tablero = "\n\n\n\n\n\nposicion 1 \t\t\tposicion 2 \t\t\tposicion 3\n\n\n\n\n\nposicion 4 \t\t\tposicion 5 \t\t\tposicion6"
        print(tablero)
    
    def movimiento_piezas():
        
        
    
        numero_de_pieza = int(input("Seleccione Pieza (1-6): "))
        posicion = int(input("Seleccione posicion (1-6): "))
        def opcion_seleccionada():
        
            opciones = {
                1:piezas_totales[0],
                2:piezas_totales[1],
                3:piezas_totales[2],
                4:piezas_totales[3],
                5:piezas_totales[4],
                6:piezas_totales[5]

            }
            return opciones[numero_de_pieza]
    
    
        if(comprobar_pieza[numero_de_pieza - 1] > 0):
            print("pieza ya insertada\n")
            return clase_rompecabezas.movimiento_piezas()
        
        else:
            if(comprobar_posicion[posicion - 1] > 0):
                print("posicion ocupada\n")
                return clase_rompecabezas.movimiento_piezas()

            else:
                rompecabezas_array[posicion - 1] = opcion_seleccionada()
                comprobar_pieza[numero_de_pieza - 1] = 1
                comprobar_posicion[posicion - 1] = 1
                
            

                print("Pieza Insertada\n")
                print(f"piezas: {comprobar_pieza[::]}")
                print(f"posicion: {comprobar_posicion[::]}\n")
                                
                
    def jugar_rompecabezas():
              
        contador = 1
        try:
            while contador < 7:
                mov = clase_rompecabezas.movimiento_piezas()
                contador +=1
        except ValueError as error:
            print("exit")
    
    def resultado():
        fig = plt.figure()
        
        a = fig.add_subplot(1, 3, 1)
        imgplot = plt.imshow(rompecabezas_array[3]);
        a.axis('off')


        a = fig.add_subplot(1, 3, 2)
        imgplot = plt.imshow(rompecabezas_array[4]);
        a.axis('off')

        a = fig.add_subplot(1, 3, 3)
        imgplot = plt.imshow(rompecabezas_array[5]);
        a.axis('off')


        a = fig.add_subplot(2, 3, 1)
        imgplot = plt.imshow(rompecabezas_array[0]);
        a.axis('off')

        a = fig.add_subplot(2, 3, 2)
        imgplot = plt.imshow(rompecabezas_array[1]);
        a.axis('off')


        a = fig.add_subplot(2, 3, 3)
        imgplot = plt.imshow(rompecabezas_array[2]);
        a.axis('off')
    


```

```python
rompecabezas_array = [0]*6
comprobar_pieza = [0]*6
comprobar_posicion = [0]*6

def funciones_rompecabezas():
    entrada = 0
    while (entrada < 5):
        print("1. Mostrar Rompecabezas\n2. Mostrar Piezas\n3. Mostrar Tablero\n4. Jugar\n5. Salir")
        entrada = int(input("Seleccione una opcion: "))
        if(entrada == 1):
            clase_rompecabezas.mostrar_rompecabezas()
        if(entrada == 2):
            clase_rompecabezas.piezas()
        if(entrada == 3):
            clase_rompecabezas.mostrar_tablero()
        if(entrada == 4):
            clase_rompecabezas.jugar_rompecabezas()
            clase_rompecabezas.resultado()
        
            
            
            
funciones_rompecabezas()
```

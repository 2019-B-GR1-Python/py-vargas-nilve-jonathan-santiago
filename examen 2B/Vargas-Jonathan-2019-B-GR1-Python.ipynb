### Ejercicio 5

```python
arreglo_random = np.random.randint(
    0, 
    50, 
    60
)
data5 = arreglo_random.reshape(10,6)
data5
colums5 = ['a','b','c','d','e','f']
df5 = pd.DataFrame(data5,columns=colums5)

def estadisticas(serie,tipo):
    if(tipo == 'mean'):
        resultado = serie.mean()
    if(tipo == 'min'):
        resultado = serie.min()
    if(tipo == 'max'):
        resultado = serie.max()
    if(tipo == 'std'):
        resultado = serie.std()
    if(tipo == 'count'):
        resultado = serie.count()
        
    return resultado

mean = estadisticas(df5['a'],'mean') + estadisticas(df5['b'],'mean') + estadisticas(df5['c'],'mean') + estadisticas(df5['d'],'mean') + estadisticas(df5['e'],'mean')+ estadisticas(df5['f'],'mean')
print (f'mean {mean}')

minimo = estadisticas(df5['a'],'min')
maximo = estadisticas(df5['a'],'min')
std = estadisticas(df5['a'],'std')
count = estadisticas(df5['a'],'std')

print (f'minimo {minimo}')
print (f'maximo {maximo}')
print (f'std {std}')
print (f'count {count}')
```

```python
arreglo_random10 = np.random.randint(
    1, 
    10, 
    60
)

data10 = arreglo_random10.reshape(10,6)

colums10 = ['a','b','c','d','e','f']

df10 = pd.DataFrame(data10,columns=colums10)
df10

def operaciones(serie,tipo):
    if(tipo == 'media'):
        resultado = serie.mean()
    if(tipo == 'mediana'):
        resultado = serie.median()
    if(tipo == 'promedio'):
        resultado = sum(serie)/float(len(serie))
  
    return resultado

mean = operaciones(df5['a'],'media') + operaciones(df5['b'],'media') + operaciones(df5['c'],'media') + operaciones(df5['d'],'media') + operaciones(df5['e'],'media')+ operaciones(df5['f'],'media')
print (f'mean {mean}')

mediana = operaciones(df5['a'],'mediana') + operaciones(df5['b'],'mediana') + operaciones(df5['c'],'mediana') + operaciones(df5['d'],'mediana') + operaciones(df5['e'],'mediana')+ operaciones(df5['f'],'mediana')
print (f'mediana {mediana}')

promedio = operaciones(df5['a'],'promedio') + operaciones(df5['b'],'promedio') + operaciones(df5['c'],'promedio') + operaciones(df5['d'],'media') + operaciones(df5['e'],'promedio')+ operaciones(df5['f'],'promedio')
print (f'promedio {promedio}')
```
